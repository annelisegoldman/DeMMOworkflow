```python
# Goal: make plots of HK count vs geochemical parameters and fit a model
# import relevant packages, etc.
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.ticker import MaxNLocator
from sklearn.linear_model import LinearRegression
import seaborn as sns
from matplotlib.pyplot import figure

# increase figure image quality 
plt.rcParams['figure.dpi'] = 300
plt.rcParams['savefig.dpi'] = 300
```

```python
#Takes file "120521_fig4.csv" (old numbering)
df = pd.read_csv(r"120521_fig4.csv")
df
```

```python
ORP_SD_X = pd.DataFrame(df['ORP_SD'])
ORP_SD_Y = pd.DataFrame(df['avg_HK_per_100'])

ORP_SD_mod = LinearRegression().fit(ORP_SD_X, ORP_SD_Y)
ORP_SD_r_sq = ORP_SD_mod.score(ORP_SD_X, ORP_SD_Y)
ORP_SD_Y_pred = ORP_SD_mod.predict(ORP_SD_X)

ax_ORP = plt.scatter(ORP_SD_X, ORP_SD_Y, color = ('#0504aa','#069af3','#4f9153','#fac205','#f97306','#980002'))
plt.plot(ORP_SD_X, ORP_SD_Y_pred, color='black')
plt.text(50, 1.0, '$r^2$ = %s'% round(ORP_SD_r_sq, 3), fontsize = 10, color = 'black')
plt.text(47, 0.5, 'ORP(mV) standard deviation', ha='center')
plt.text(22, 0.95, 'average #HKs/100 genes/site', va='center', rotation='vertical')

plt.savefig('120521_ORP_SD_fig4.png', orientation='landscape', bbox_inches = "tight")
```
